```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# 원소에 접근할 때 (access, 값을 집어넣고나 꺼내올 때)
a = np.zeros((10, 10))
# 튜플속에 각 어레이별 앤덱스를 넣는다.
a = np.arange(9).reshape(3, -3)
a[1, 2]
#7을 10으로 바꾸려면???
```

```python
#인덱싱.. 단일 원소에 접근
#튜프로 접근... 리스트(x) 넘파이 행렬(x)
a = np.zeros((10, 10))
a[(0, 0)] = 1
a[(1, 1)] = 1
# 튜플은 괄호 생략가능
a[2, 3] = 1
a[4, 4] = -1
a[4, 5] = -2
a
```

```python
# 10x10 크기의 행렬 - 이미지를 담는 그릇
# python에서 numpy 행렬을 이미지 저장 포맷으로 활용
plt.imshow(a)
# float 이고 채널이 없는 gray 영상
# color라면.. 채널이 있어야 되는데? (10, 10, 3)
# gray영상은 plt에서 출력할 때 MinMaxScaling (-2, 1) -> (0, 255)
# -> 0은 검은색, 255는 흰색인데.. 컬러맵이 적용된 것
# 0-> 보라색 1-> 약간보라색.... 255->노란색
#            .. 모든 gray레벨 별 색상 매칭
# 컬러맵 활용하는 이유 - 흑백으로만 되면 값의 미세한 차이가 안 보이기 때문에
```

```python
# reshape은 크기를 바꾸는 연산인데
# reshape은... element의 개수가 똑같다면 얼마든지 shape 변환 가능
# 데이터는 그대로 놔두고 헤더 부분만 변경
# numpy는 두 부분으로 나눔, header, data
a = np.arange(6).reshape(2, 3) # 그냥 버려지고 메모리(RAM)에 보관화지 않음. garbage collection
a
```

```python
a.reshape(6).reshape(3, 2) # a는 안 바뀌고 바뀐 것을 리턴
```
